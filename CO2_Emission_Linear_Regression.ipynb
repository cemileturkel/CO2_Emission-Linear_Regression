```python
# import libraries

import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
```

```python
# Data Frame = df

df = pd.read_csv('/CO2_emission.csv')
selected_columns = df[['Fuel_Consumption_in_City_Hwy(L/100 km)', 'CO2_Emissions']]
print(selected_columns)
```

```python
# draw a table

plt.scatter(df ['Fuel_Consumption_in_City_Hwy(L/100 km)'], df ['CO2_Emissions'], color="Red")
plt.xlabel('FUEL CONSUMPTION_HWY')
plt.ylabel('CO2 EMISSION')
plt.title('FUEL CONSUMPTION-CO2 EMISSION Graph')
plt.show()
```

```python
# define properties and target variable

x = df[['Fuel_Consumption_in_City_Hwy(L/100 km)']]
y = df['CO2_Emissions']
```

```python
# train model

model = LinearRegression()
model.fit(x, y)
```

```python
# You can make some predictions to verify that the model is trained

y_pred = model.predict(x)
y_pred
```

```python
# draw on graph

plt.scatter(x, y, color="Red")
plt.plot(x, y_pred, color="Blue")
plt.xlabel('FUEL CONSUMPTION_HWY')
plt.ylabel('CO2 EMISSION')
plt.title('FUEL CONSUMPTION-CO2 EMISSION Graph')
plt.show()
```

```python
# find a and b

a = model.coef_
b = model.intercept_
print(a,b)
```

```python
# You can calculate R score to evaluate the performance of the model

r2 = r2_score(y, y_pred)
print(f'R Score: {r2}')
```

```python
# Estimate the CO2 emissions of a vehicle with FUEL CONSUMPTION_HWY '15'

fuel_consumption_hwy = pd.DataFrame({'Fuel_Consumption_in_City_Hwy(L/100 km)': [15]})
predicted_co2_emission = model.predict(fuel_consumption_hwy)

print(f'FUEL CONSUMPTION_HWY 15 olan bir aracn tahmin edilen CO2 emisyonu: {predicted_co2_emission[0]}')
```
